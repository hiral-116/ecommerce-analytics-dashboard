```python
df['full_name'] = df['first_name'] + ' ' + df['last_name']
display(df.head())
```

```python
columns = df.columns.tolist()
columns.remove('first_name')
columns.remove('last_name')
columns.insert(1, columns.pop(columns.index('full_name'))) # Move 'full_name' after 'customer_id'
df = df[columns]
display(df.head())
```

```python
# Define the date columns to convert
date_columns = ['signup_date', 'order_date', 'review_date']

# Convert date columns to datetime objects with a specific format
for col in date_columns:
    # Using errors='coerce' will turn unparseable dates into NaT
    df[col] = pd.to_datetime(df[col], format='%d-%m-%Y', errors='coerce')

df.info()
```

```python
df["total_amount"] = df["quantity"] * df["unit_price"]
columns = df.columns.tolist()
columns.insert(11, columns.pop(columns.index('total_amount'))) # Move 'full_name' after 'customer_id'
df = df[columns]
display(df.head())
```
